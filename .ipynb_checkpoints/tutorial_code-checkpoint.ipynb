```python
cap = cv2.VideoCapture(0)
#video codeing for save
fourcc = cv2.VideoWriter_fourcc(*'XVID')
#here we make our saving object cv2.VideoWriter('direction',coding,20.0, (size))
out = cv2.VideoWriter('output.avi',fourcc,20.0, (640,480))
while True:
    ret, frame = cap.read()
    #this way we convert our frames to gray color
    #gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    out.write(frame)
    cv2.imshow('frame',frame)
    #cv2.imshow('gray',gray)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread("lena.jpg")
#drawing line cv2.line(img, (start_point) , (end_point), (color_rgb),thiknes)
cv2.line(img, (0,0) , (150,150), (255,255,255),6)
#drawing rectangle cv2.rectangle (img, (start_point) , (end_point), (color_rgb),thiknes)
cv2.rectangle (img, (0,0) , (150,150), (0,255,255),6)
#drawing cirlcle cv2.circle(img,(center_point),redius,(rgb_color),thikness)
#if u set thickness to -1 it will fill the shape
cv2.circle(img,(100,150),60,(255,0,255),3)
#drawing any shape first u need to make an arry of points
pts = np.array([[10,5],[40,30],[90,80]],np.int32)
#if needed u need to reshape numpy array
#reshape code is pts = pts.reshape((-1,1,2))
#now point drawing code cv2.polylines(img,[array_of_points],True_to_connect_points,(color_rgb),thickness)

cv2.polylines(img,[pts],True,(0,255,255),3)

#if u want to write something u need to do this
#first define the font
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,"Hello World!",(0,130),font,1,(200,200,100),5,cv2.LINE_AA)

cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
